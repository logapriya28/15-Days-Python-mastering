## a. for loop
- for loop is used to iterate over a sequence like list,tuple,string or other iterable objects.

## c.loop control statements
- Python provides loop control statements to alter the behaviour of loops.

     ## a. break
     - This terminates the loop prematurely when a certain condition is met.
     
        ## b.continue
         - It skips the rest of the code inside a loop for the current iteration and continues to next iteration.
         
           ## c.pass
           - Used as  placeholder when a statement is syntactically required but you don't want any action to occur.

## Write a program that checks if a given number is positive, negative, or zero.

```python
number=int(input("Enter a number to check :"))

if number ==0:
    print("The number is zero")
elif number>0:
    print("The number is positive")
else:
    print("The number is negative")
```

## Create a program that takes a year as input and checks if it is a leap year or not

```python
year=int(input("Enter the year :"))

if year%100==0:
    if year%400==0:
        print("The year you entered is a leap year")
    else:
        print("The year you entered is not a leap year")
elif year %4==0:
    print("the year you entered is a leap year")
else:
        print("The year you entered is not a leap year")
```

##  Given a list of integers, find all the even numbers and store them in a new list

```python
list_num=input("Enter the numbers seperated by comma :").split(",")

#converting string to integer datatype
list_num=[int(num) for num in list_num]


#creating an empty list to store the even numbers
new_list=[]  

for i in list_num:
    if int(i)%2==0:
        new_list.append(i)
print(new_list)
```

## Write a Python program to check if a given number is a prime number

```python
num=93751

if num<=1:
    print("The number is not prime")
else:    
    for i in range(2,num):
        if num%i==0:
            print("The number is non prime")
            break
    else:
        print("the number is prime")
```

##  Create a program that generates the Fibonacci sequence  up to a given number of terms

```python
num1=0
num2=1
new_num=num2

count=1

fib_num=int(input("Enter the number :"))

while count<=fib_num:
    print(new_num)
    count+=1
    num1,num2=num2,new_num
    new_num=num1+num2

print()
```

```python
table=int(input("Enter the multiplication table you want :"))
numbers=int(input("Enter the range of numbers for the table :"))

for i in range(numbers+1):
    print(f"{i}*{table} = {i*table}")
```

## Write a program that calculates the factorial of a given number

```python
n=int(input("enter a number :"))

def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)    #recursion

num=factorial(n)
print("The factorial is ",num)
```

## Create a loop that prints all prime numbers between 1 and 50

```python
for num in range(1,50):
    prime=True
    
    for i in range(2,num):
        if num %i==0:
            prime=False
        
    if prime:
        print(num)
        
```
