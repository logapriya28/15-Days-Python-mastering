- In Python, lists and tuples are both sequence data types used to store collections of items. They have some similarities but also distinct differences. Let's explore them:

**Lists:**
- Lists are ordered collections of items.
- They are mutable, meaning you can modify the elements after creation.
- Lists are defined using square brackets [].
- You can have heterogeneous data types within a single list.
- Lists support operations like appending, inserting, removing, and slicing.

**Tuples:**
- Tuples are ordered collections of items, just like lists.
- They are immutable, meaning you cannot change the elements after creation.
- Tuples are defined using parentheses ().
- Tuples are often used for fixed collections of items where immutability is desired.
- Tuples can also have heterogeneous data types.

**Operations common to both lists and tuples:**
- Indexing: Accessing elements using their position in the sequence.
- Slicing: Extracting a subset of elements from the sequence.
- Concatenation: Combining two sequences into one.
- Membership: Checking if an element is present in the sequence.
- Length: Finding the number of elements in the sequence.

**Differences:**
- Lists use square brackets [], while tuples use parentheses ().
- Lists are mutable, while tuples are immutable.
- Lists have more built-in methods for modification, like append(), insert(), remove(), etc., which are not available for tuples.
- Tuples are often used in scenarios where immutability and integrity are desired, like representing fixed collections such as coordinates or database records.

## Write a Python program to count the occurrences of each element in a given list

```python
def count_occurrences(input_list):
    element_counts = {}

    for element in input_list:
        if element in element_counts:
            element_counts[element] += 1
        else:
            element_counts[element] = 1
    
    for element, count in element_counts.items():
        print(f"{element}: {count}")

input_list = [1, 2, 3, 1, 2, 3, 4, 1, 2, 1]
    
count_occurrences(input_list)
```

## Given a list of names, remove all duplicate names and print the unique names

```python
def remove_duplicates(names):
    unique_names = []
    for name in names:
        if name not in unique_names:
            unique_names.append(name)
    return unique_names

names_list = ["John", "Doe", "Jane", "Doe", "John", "Smith", "Jane"]
unique_names = remove_duplicates(names_list)
print("Unique names:")
for name in unique_names:
    print(name)
```

## Create a function that takes a list of strings and returns the list sorted by the length of the strings

```python
def sort_strings_by_length(string_list):
    return sorted(string_list, key=len)

string_list = ["banana", "apple", "kiwi", "strawberry", "pineapple"]
sorted_list = sort_strings_by_length(string_list)
print(sorted_list)
```

## Write a program that checks if a given list is sorted in ascending order

```python
def is_sorted_ascending(lst):
    if len(lst) <= 1:
        return True
    
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            return False  
    return True  
list1 = [1, 2, 3, 4, 5]

print("Is list1 sorted in ascending order ?", is_sorted_ascending(list1)) 
```

## Implement a function that takes two lists and returns their union (all unique elements from both lists)

```python
def union_lists(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    return list(set1.union(set2))

list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

union_result = union_lists(list1, list2)
print("Union of list1 and list2:", union_result)
```
