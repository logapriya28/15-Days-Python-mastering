- **String manipulation** in Python involves performing various operations on strings such as concatenation, slicing, searching, replacing, splitting, and formatting. 
- Python provides a rich set of built-in functions and methods for working with strings.

















- **Concatenation:** Combining two or more strings together.
- **Slicing:** Extracting a portion of a string.
- **Searching:** Finding substrings within a string.
- **Replacing:** Replacing parts of a string with another substring.
- **Splitting:** Splitting a string into a list of substrings.
- **Joining:** Joining a list of strings into a single string.
- **Case Conversion:** Converting the case of characters in a string.
- **Stripping:** Removing leading and trailing whitespace from a string.
- **Formatting:** Formatting strings using placeholders or format() method.

## Create a program that checks if a given string is a palindrome.

```python
string='hey world'

if string[::-1]==string:
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

 ## Write a function to count the number of vowels in a given string

```python
def count_vowels(string):
    vowels = "aeiouAEIOU"
    count = 0
    for char in string:
        if char in vowels:
            count += 1
    return count

input_string = "Hello, how are you?"
print("Number of vowels:", count_vowels(input_string))
```

## Create a function to reverse a given string


```python
def reverse(string):
    return string[::-1]

string=input('enter a string :')
output=reverse(string)
print('The reversed string is :',output)
```

## Implement a function that checks if a given string is a pangram (contains all letters of the alphabet

```python
def pangram(string):
    string=string.lower()
    alphabet=set("abcdefghijklmnopqrstuvwxyz")
    return set(string)>=alphabet

string=input("Enter a Sentence :")
if pangram(string):
    print("The sentence is a pangram")
else:
    print("The sentence is not a Palgram")
```

##  Given a string, write a function to remove all vowels from it and return the modified string'

```python
string='unicorntemple'
vowels=set("aeiou")
new_string=""
for i in string:
    if i not in vowels:
        new_string+=i
        
print("The modified string is :",new_string)   
```

## Write a Python program to find the length of the longest word in a sentence

```python
input_string=input("Enter a sentence :")
words=input_string.split()
word_count=0
for i in words:
    if len(i)>word_count:
        word_count=len(i)
print('The length  of the longest word is :',word_count) 
```

## Given a list of names, count the number of names that start with a vowel

```python
list_name=['abi','charley','wilson','ilaa','edwin','diana']
vowels=set("aeiouAEIOU")
count=0

for i in list_name:
    if i[0] in vowels:
        count+=1
    
print(count)
```

## Write a function to remove all duplicate characters from a given string

```python
def remove_duplicates(input_string):
    unique_chars = set()
    
    result = ''
    
    for char in input_string:
        if char not in unique_chars:
            unique_chars.add(char)
            result += char
    
    return result

input_string = "hello"
print("String after removing duplicates:", remove_duplicates(input_string))
```
