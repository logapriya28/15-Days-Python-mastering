## Variables

- Variables in Python are used to store data values. 
- They are essentially reserved memory locations to store values.
- You declare variables in Python by simply assigning a value to them using the equal (=) sign.
- Variable names can contain letters (a-z, A-Z), digits (0-9), and underscores (_).
- Variable names cannot start with a digit.
- Variable names are case-sensitive (myVar and myvar are different variables).
- Variable names cannot be a reserved keyword (e.g., if, while, for, etc.).

## Datatypes

* Python is dynamically typed, meaning you don't need to declare the data type of a variable explicitly. 
Python infers the data type based on the value assigned to it. Here are some common data types in Python:

**Numeric Types:**
- int: Integer values, e.g., 5, 100, -10.
- float: Floating-point values, e.g., 3.14, 2.71828, 0.0.

**Sequence Types:**
- str: String values, e.g., "hello", "Python", "123".
- list: Ordered collection of items, e.g., [1, 2, 3], ['a', 'b', 'c'].
- tuple: Immutable ordered collection of items, e.g., (1, 2, 3), ('a', 'b', 'c').

**Mapping Type:**
- dict: Collection of key-value pairs, e.g., {'name': 'John', 'age': 25}.

**Set Types:**
- set: Unordered collection of unique items, e.g., {1, 2, 3}, {'a', 'b', 'c'}.
- frozenset: Immutable version of set.

**Boolean Type:**
- bool: Represents truth values True or False.

**None Type:**
- NoneType: Represents the absence of a value, typically used as the default return value of functions that do not explicitly return anything.

## 1.Given a list of numbers, find the sum and average

```python
num_list=[23,45,90,78,12,78,34,90,22]
number=0

for i in num_list:
    number+=i
    

avg=len(num_list)
average=number/avg
round_avg=round(average,2)    #rounding off the average obtained

print("The sum of the list is ",number)
print("The average of the list is ",round_avg)
```

## 3. Implement a program that checks if a given string is a palindrome

```python
check=input("Enter a string :")

# When upper case strings are given,there will be a chance of miscalculation hence all words are converted to lower case
check_palindrome=check.lower() 

if check_palindrome==check_palindrome[::-1]:
    print(f"{check} is a palindrome")
else:
    print(f"{check} is not a palindrome")
    
```

## 4.Create a function to reverse a given string

```python
def reverse_string(input_string):
    return input_string[::-1]

input_string=input("Enter a string: ")

print("The string you entered :"+input_string)
output=reverse_string(input_string)

print("The reversed string :",output)
```

## 6.Write a Python program to check if a given string is a pangram (contains all letters of the alphabet)

```python
def pangram(string):
    string=string.lower()
    alphabet=set("abcdefghijklmnopqrstuvwxyz")
    return set(string)>=alphabet

string=input("Enter a Sentence :")
if pangram(string):
    print("The sentence is a pangram")
else:
    print("The sentence is not a Palgram")
```

## 8.Implement a program that converts a given number of minutes into hours and minutes

```python
def hours(minutes):
    hour=minutes//60
    remaining_minutes=minutes%60
    return hour,remaining_minutes

minutes=int(input("Enter the minutes to convert :"))
hour,remaining_minutes=hours(minutes)

print(f"{hour} hr and {remaining_minutes} mins")
```

## 9.Create a function to count the number of vowels in a given string

```python
string=input(" Enter a string :")
string=string.lower()

list_vowels=[] #creating an empty list to store the vowels
vowel=set("aeiou")
    
for i in string:
   
    if i in vowel:
        list_vowels.append(i)
    else:
        continue
        
print(len(list_vowels))
        
        
```

## 10.Write a program to check if a number is prime.

```python
num=int(input("Enter a number :"))

if num==1:
    print("It is not a prime number")
else:
    for i in range(2,num):
        if num%i==0:
            print("It is not a prime number")
            break
    else:
        print("It is a prime number")
```
